```python
import random


def GetMoves(self):
    """ Get all possible moves from this state.
    """
    sfreeagents = self.freeagents
    
    if len(self.turns) == 0: return []
    if len([self.rosters[self.turns[0]]][0]) == num_rounds: return []
    l = len(sfreeagents) - (zongshu - totalp)
    moves = order[num_rounds - ceil(l/num_competitors)]
    if moves > len(sfreeagents):
        moves = len(sfreeagents)
#    moves = order[len([self.rosters[self.turns[0]]][0])]
#    n = len([self.rosters[self.turns[0]]][0])
#    moves = range(n,13, 1)

# return all possible players' range
    return list(range(moves))

DraftState.GetMoves = GetMoves


def DoMove(self, move):
    """ Update a state by carrying out the given move.
        Must update playerJustMoved.
    """
#   find the player
    #    player = next(p for p in self.freeagents if p.position == move)
    player = self.freeagents[move]
    self.freeagents.remove(player)
    rosterId = self.turns.pop(0)
    self.rosters[rosterId].append(player)
    self.playerJustMoved = rosterId    #which team just moved

DraftState.DoMove = DoMove

def Clone(self):
    """ Create a deep clone of this game state.
    """
    rosters = list(map(lambda r: r[:], self.rosters))
    st = DraftState(rosters, self.turns[:], self.freeagents[:],
            self.playerJustMoved)
    return st
DraftState.Clone = Clone
```

```python
from math import *
import random



class Node:
    """ A node in the game tree. Note wins is always from the viewpoint of playerJustMoved.
        Crashes if state not specified.
    """
    def __init__(self, move = None, parent = None, state = None, pre = None):
        self.move = move # the move that got us to this node - "None" for the root node
        self.parentNode = parent # "None" for the root node
        self.pre = pre
        self.childNodes = []
        self.wins = 0
        self.visits = 0
        self.untriedMoves = state.GetMoves() # future child nodes
        self.playerJustMoved = state.playerJustMoved # the only part of the state that the Node needs later

    def UCTSelectChild(self):
        """ Use the UCB1 formula to select a child node. often a constant UCTK is applied so we have
            lambda c: c.wins/c.visits + UCTK * sqrt(2*log(self.visits)/c.visits to vary the amount of
            exploration versus exploitation.
        """
        UCTK = 100/np.sqrt(2)*0.7
        s = sorted(self.childNodes, key = lambda c: c.wins/c.visits + UCTK * sqrt(2*log(self.visits)/c.visits))[-1]
        return s

    def AddChild(self, m, s, n):
        """ Remove m from untriedMoves and add a new child node for this move.
            Return the added child node
        """
        n = Node(move = m, parent = self, state = s, pre = n)
        self.untriedMoves.remove(m)
        self.childNodes.append(n)
        return n

    def Update(self, result):
        """ Update this node - one additional visit and result additional wins. result must be from the viewpoint of playerJustmoved.
        """
        self.visits += 1
        self.wins += result
        
def UCT(rootstate, itermax, hz, verbose = False):
    """ Conduct a UCT search for itermax iterations starting from rootstate.
        Return the best move from the rootstate.
    """
    choice = random.choice
    rootnode = Node(state = rootstate)
#    cplay = rootnode.playerJustMoved
    
    num_rounds = 13
    num_competitors = 16
    standard = np.array(range(0, 156, 12)) 
    hz = hz
    
    for i in range(itermax):
        node = rootnode
        state = rootstate.Clone()
        Dmove = state.DoMove
        Gmove = state.GetMoves
        sfreeagents = state.freeagents
        # Select
        while node.untriedMoves == [] and node.childNodes != []: # node is fully expanded and non-terminal
            node = node.UCTSelectChild()
            Dmove(node.move)
        # Expand
        if node.untriedMoves != []: # if we can expand (i.e. state/node is non-terminal)
#            m = Gmove() - node.untriedMoves
#            n = choice(range(order[m]))  # n stands for player, m stands for place
            m = choice(node.untriedMoves)
            l = len(sfreeagents) - (zongshu - totalp)
            length = num_rounds - ceil(l/num_competitors)
            n = sfreeagents[m].pre
            if (n > standard[length]):
                Dmove(m)
                node = node.AddChild(m, state, n) # add child and descend tree
            

        # Rollout - this can often be made orders of magnitude quicker using a state.GetRandomMove() function
        while Gmove() != []: # while state is non-terminal
            m = choice(Gmove())
            l = len(sfreeagents) - (zongshu - totalp)
            length = num_rounds - ceil(l/num_competitors)
            n = sfreeagents[m].pre
            if (n > standard[length]):
                Dmove(m)
        
        result.extend([[state.rosters[i], state.GetResult(i)] for i in range (num_competitors) if state.GetResult(i) > 113])
#        result.extend([[state.rosters[i], state.GetResult(i)] for i in range (num_competitors) if state.GetResult(i) > 128])
        
        # Backpropagate
        while node != None: # backpropagate from the expanded node and work back to the root node
            node.Update(state.GetResult(node.playerJustMoved)) # state is terminal. Update node with result from POV of node.playerJustMoved
            node = node.parentNode
    bnode = sorted(rootnode.childNodes, key = lambda c: c.visits)[-1]        
    return bnode.move, bnode.pre, rootnode # return the move that was most visited

```
